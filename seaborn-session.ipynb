# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Seaborn is built on top of matplotlib, but it offers several advantages:

- **High-level interface**: Seaborn provides a simpler and more intuitive way to create complex statistical graphics, such as heatmaps, box plots, and scatter plots.
- **Attractive default styles**: Seaborn has a built-in set of attractive default styles that make your plots look professional and visually appealing.
- **Integration with pandas**: Seaborn is designed to work seamlessly with pandas data structures, making it easy to create visualizations from datasets.

## Important Features of Seaborn
1. Built in themes for styling matplotlib graphics
2. Visualizing univariate and bivariate data
3. Fitting in and visualizing linear regression models
4. Plotting statistical time series data
5. Seaborn works well with NumPy and Pandas data structures
6. It comes with built in themes for styling Matplotlib graphics

### Seaborn Plotting Functions
scatter plot using `lmplot()`<br>

The lineplot (lmplot) is one of the most basic plots. It shows a line on a 2 dimensional plane. You can plot it with seaborn or matlotlib depending on your preference.

Two ways
1. Recommended way is to pass DataFrame to the data= argument, and pass column names to the axes arguments, x= and y=.<br>
`sns.lmplot(x='cylinders', y='horsepower', data=df)`
2. Pass in Series of data to the axes arguments.<br>
`sns.lmplot(x=df.cylinders, y=df.horsepower)`

```python
sns.lmplot(x='weight', y='horsepower', data=df)
```

The diagonal line is regression line<br>
we can remove regression line by `fit_reg=False`<br>
Give colors to plots by `hue='Stage'`

```python
sns.lmplot(x='weight', y='horsepower', data=df,
           fit_reg=False, # No regression line
           hue='origin',)
```

Seaborn is high level interface to Matplotlib. So Matplotlib is required at times.<br>
use `xlim` and `ylim` functions of Matplotlib

```python
# Plot using Seaborn
sns.lmplot(x='weight', y='horsepower', data=df,
           fit_reg=False, hue='origin')
 
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

### Violin Plot
Show distribution through the tickness of violen

```python
sns.violinplot(x='cylinders', y='horsepower', data=df)
plt.show()
```

A violin plot is a statistical chart used to visualize the distribution of numerical data, particularly when comparing distributions across different groups. 

It combines the features of a box plot and a kernel density plot, providing a more detailed view of the data's shape and spread than a traditional box plot. 

The width of the violin represents the density of data points at a given value, with wider areas indicating higher density. 


### Color Palettes

### Joint Distribution Plot

A Joint Distribution Plot in Seaborn (created using sns.jointplot()) is a multi-component visualization that shows:

1. Main plot (center): Usually a scatter plot showing the relationship between two variables

2. Marginal distributions (top and right): Shows the distribution of each individual variable

    * Top: Distribution of the x-variable

    * Right: Distribution of the y-variable

```python
sns.jointplot(x='weight', y='displacement', data=df)
```

### Pairplot

A pairplot in Seaborn is a powerful visualization tool that creates a grid of plots showing relationships between multiple variables in your dataset. It's one of the most comprehensive ways to explore multivariate relationships at once.

**What is a Pairplot?**

A pairplot creates a matrix of plots where:

* Each variable is plotted against every other variable in scatter plots (in the off-diagonal cells)

* The diagonal shows the distribution of each variable (typically as histograms or KDE plots)

* All of this is displayed in a single figure with a grid layout

```python
tips.sample(6)
```

```python
plt.figure(figsize=(3, 3))
sns.pairplot(tips)
```
